## Small exploration of the dataset

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('train.csv')
df.head(10)
```

```python
df.info()
```

```python
# Counting missing values 
missing_values = df.isnull().sum()
missing_values_percentage = (missing_values / len(df)) * 100
missing_values_percentage
```

```python
df_shape = df.shape
df_shape
```

```python
unique_label_groups = df['label_group'].nunique()
unique_label_groups
```

```python
# Checking the distribution of the number of products per label_group
df['label_group'].value_counts().hist(bins=50)
plt.title('Distribution of Number of Products per Label Group')
plt.xlabel('Number of Products')
plt.ylabel('Number of Label Groups')
plt.show()

# Checking the top 10 most common product titles
df['title'].value_counts()[:10].plot(kind='barh')
plt.title('Top 10 Most Common Product Titles')
plt.xlabel('Number of Products')
plt.ylabel('Product Title')
plt.show()

# Checking the distribution of the length of product titles
df['title'].apply(lambda x: len(x)).hist(bins=50)
plt.title('Distribution of Length of Product Titles')
plt.xlabel('Length of Title')
plt.ylabel('Number of Products')
plt.show()
```
